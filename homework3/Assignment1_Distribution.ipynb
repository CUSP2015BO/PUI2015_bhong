```python
from __future__ import  print_function
import pylab as pl


#magic command
%pylab inline
```

### Making 100 different sizes

```python
sizes=range(10, 2000, 19)
print (sizes)
```

## Normal (Gaussian) Distribution

```python
# Gaussian (Normal) Disribution

# Make data frame to put in result of distribution
distributions={}
distributions ['normal'] = {}

# *****Shuld I make mean and sd randomly?
normalmean=100
normalsd=sqrt(100)

for n in sizes:
    distributions['normal'][n]=np.random.normal(normalmean, normalsd, size=n)
```

```python
means = {}
means['normal'] = {}
sd = {}
sd['normal'] = {}

fig_mu_N = pl.figure(figsize=(10,10))
fig_sd_N = pl.figure(figsize=(10,10))
axnormal_mu_n = fig_mu_N.add_subplot(221)
axnormal_sd_n =  fig_sd_N.add_subplot(222)

for nn in distributions['normal'].iterkeys():
    means['normal'][nn] = distributions['normal'][nn].mean()
    sd['normal'][nn] = distributions['normal'][nn].std()
    #and plot it
    axnormal_mu_n.plot(nn, means['normal'][nn], 'o')
    axnormal_mu_n.set_xlabel('sample size')
    axnormal_mu_n.set_ylabel('sample mean')
    axnormal_mu_n.set_title('Scatter plot of sample means and sample size')
    axnormal_mu_n.plot([min(sizes), max(sizes)],[normalmean,normalmean], 'k')
    
    axnormal_sd_n.plot(nn, sd['normal'][nn], 'o')
    axnormal_sd_n.set_xlabel('sample size')
    axnormal_sd_n.set_ylabel('sample standard deviation')
    axnormal_sd_n.set_title('Scatter plot of sample sd and sample size')
    axnormal_sd_n.plot([min(sizes), max(sizes)], [normalsd,normalsd], 'k')
```

```python
# Plot the the histogram of the mean of all samples (Normal Distribution)

normalmeans=[]
for n in means['normal']:
    normalmeans.append(means['normal'][n])
pl.figure(figsize=(10,10))
pl.hist(normalmeans,color='steelBlue',bins=30, alpha=0.5)
pl.xlabel('sample mean')
pl.ylabel('N')
pl.title('Histogram of sample mean')

```

##Chi-square Distribuion

```python
# Chi-square Disribution

# Make data frame to put in result of distribution
distributions ['chi'] = {}

#the chi sq distribution takes the number of degrees of freedom, which luckily is the same as the mean!
chidf = 100

for n in sizes:
    distributions['chi'][n]=np.random.chisquare(chidf, size=n)
```

```python
means['chi'] = {}
sd['chi'] = {}

fig_mu_C = pl.figure(figsize=(10,10))
fig_sd_C = pl.figure(figsize=(10,10))
axchisq_mu_c = fig_mu_C.add_subplot(221)
axchisq_sd_c =  fig_sd_C.add_subplot(222)

for nn in distributions['chi'].iterkeys():
    means['chi'][nn] = distributions['chi'][nn].mean()
    sd['chi'][nn] = distributions['chi'][nn].std()
    #and plot it
    axchisq_mu_c.plot(nn, means['chi'][nn], 'o')
    axchisq_mu_c.set_xlabel('sample size')
    axchisq_mu_c.set_ylabel('sample mean')
    axchisq_mu_c.set_title('Scatter plot of sample means and sample size')
    axchisq_mu_c.plot([min(sizes), max(sizes)],[chidf,chidf], 'k')
    
    axchisq_sd_c.plot(nn, sd['chi'][nn], 'o')
    axchisq_sd_c.set_xlabel('sample size')
    axchisq_sd_c.set_ylabel('sample standard deviation')
    axchisq_sd_c.set_title('Scatter plot of sample sd and sample size')
    #axchisq_sd_c.plot([min(sizes), max(sizes)], [chidf,chidf], 'k')
```

```python
# Plot the the histogram of the mean of all samples (Chi-sqaure Distribution)
chimeans=[]

for n in means['chi']:
    chimeans.append(means['chi'][n])
pl.figure(figsize=(10,10))
pl.hist(chimeans,color='steelBlue',bins=30, alpha=0.5)
pl.xlabel('sample mean')
pl.ylabel('N')
pl.title('Histogram of sample mean')
```
